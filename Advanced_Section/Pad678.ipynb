```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
```

```python
df_us
```

```python
df_us_pivot = df_us.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
```

```python
df_us_pivot.reset_index(inplace=True)
```

```python
df_us_pivot['month_number'] = pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
```

```python
df_us_pivot
```

```python
df_us_pivot.sort_values('month_number',inplace=True)
```

```python
df_us_pivot.set_index(['job_posted_month'],inplace=True)
```

```python
df_us_pivot
```

```python
top_5=df_us['job_title_short'].value_counts().head()
```

```python
top_5
```

```python
top_5 = top_5.index.tolist()
```

```python
df_us_pivot[top_5].plot(kind='line')
plt.title('Job Posting by Month by Profession')
plt.xlabel('Month')
plt.ylabel('# of Postings')
plt.legend()
```

```python
df_us_software_pivot = pd.read_csv("https://lukeb.co/software_csv",index_col="job_posted_month")
```

```python
df_us_merged = df_us_pivot.merge(df_us_software_pivot,on='job_posted_month')
```

```python
top_5 = df_us_merged.sum().sort_values(ascending=False).head().index.to_list()

top_5
```

```python
df_us_merged[top_5].plot(kind='line')
plt.title('Data and Programming US Job Postings by Month')
plt.xlabel('Month (2023)')
plt.ylabel('# of Postings')
plt.ylim(2000,22000)
plt.legend()
```
