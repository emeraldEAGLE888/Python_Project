Exploratory Data Analysis

```python
import pandas as pd
import ast
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
my_df=df[(df['job_title_short']=='Data Analyst') & (df['job_country'].isin(['United States','Canada']))]
```

```python
df_plot = my_df['job_location'].value_counts().head(10).to_frame()
```

# Most Common Locations for Data Analysts in US/Canada

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='rocket_r',legend=False)
sns.despine()
plt.title('Jobs per Location for US/Canada Data Analysts')
plt.ylabel('')
plt.xlabel('Count')
```

# Analysis of Benefits for Data Analysts in US/Canada

```python
fig, ax = plt.subplots(1, 3)
fig.set_size_inches((10,4))
my_df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%',ax=ax[0],title='Work from Home %',ylabel='')
my_df['job_no_degree_mention'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%',ax=ax[1], title='No Degree %',ylabel='')
my_df['job_health_insurance'].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%',ax=ax[2],title='Health Insurance %')
plt.ylabel('')
plt.xlabel('')
fig.tight_layout()
plt.show()
```

# Most Popular Companies for Data Analysts in US/Canada

```python
df_plot_2 = my_df['company_name'].value_counts().head(10).to_frame()
sns.barplot(data=df_plot_2,x='count',y='company_name',hue='count',palette='rocket_r',legend=False)
sns.despine()
plt.title('Jobs per Location for US/Canada Data Analysts')
plt.ylabel('')
plt.xlabel('Count')
```
